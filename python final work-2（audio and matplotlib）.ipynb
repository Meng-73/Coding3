## Read a piece of audio and use matplotlib to draw a histogram of this audio



```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
```

```python



wavFile = "music01.wav"


data, samplerate = sf.read(wavFile)

n = len(data)  
Fs = samplerate  


ch1, ch2 = data.transpose()


time_axis = np.linspace(0, n / Fs, n, endpoint=False)
sound_axis = ch1 


plt.plot(time_axis, sound_axis)
plt.show()
```
